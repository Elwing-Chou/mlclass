```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris["data"], columns = iris["feature_names"])
df["target"] = iris["target"]
df
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, target_train, target_test = train_test_split(df.drop(["target"], axis = 1),
                                                                    df["target"],
                                                                    test_size = 0.1)
```

```python
from sklearn.cluster import KMeans
clu = KMeans(n_clusters = 3)
clu.fit(data_train)
pre = clu.predict(data_test)
```

```python
print("預設標籤", list(pre))
print("正確標籤", list(target_test))
from sklearn.metrics import accuracy_score
print("正確率", accuracy_score(pre, target_test) * 100, "%")
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
%matplotlib inline
# 累積畫圖的東西
xs = []
ys = []
# range(2, 6) = [2, 3, 4, 5]
total_train = df.drop(["target"], axis = 1)
for k in range(2, 6):
    clu = KMeans(n_clusters = k)
    clu.fit(total_train)
    s = silhouette_score(total_train, clu.labels_)
    xs.append(k)
    ys.append(s)
print("x座標", xs)
print("y座標", ys)
plt.plot(xs, ys)
```
