```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding = "utf-8")
train_df
```

```python
test_df = pd.read_csv("test.csv", encoding = "utf-8")
test_df
```

```python
train_df.isnull()
```

```python
train_df.isnull().sum()
```

```python
train_df = train_df.fillna(train_df.median())
train_df.isnull().sum()
```

```python
train_df["Embarked"].value_counts()
```

```python
often_value = train_df["Embarked"].value_counts().idxmax()
train_df["Embarked"] = train_df["Embarked"].fillna(often_value)
train_df.isnull().sum()
```

```python
train_df = train_df.drop(["PassengerId", "Ticket", "Cabin"], axis = 1)
train_df
```

```python
# é¡žåˆ¥ä¸€å®šè¦è½‰æ›æˆOne-hot Encoding
# è½‰æ›å®Œä½ æ‰å¯ä»¥æŠŠåŽŸæœ¬çš„ä¸ŸæŽ‰
dummy = pd.get_dummies(train_df["Embarked"])
train_df = pd.concat([train_df, dummy], axis = 1)
dummy = pd.get_dummies(train_df["Sex"])
train_df = pd.concat([train_df, dummy], axis = 1)
train_df = train_df.drop(["Sex", "Embarked"], axis = 1)
train_df
```

```python
s = train_df["Name"].str.split(",", expand = True)[1]
s = s.str.split(" ", expand = True)[1]
pd.crosstab(s, train_df["Survived"]).T.style.background_gradient(cmap="autumn")
```

```python
def filter_name(data):
    if data == "Mr.":
        return "Mr"
    elif data == "Mrs.":
        return "Mrs"
    elif data == "Miss.":
        return "Miss"
    else:
        return "Undefined"
train_df["Name"] = s.apply(filter_name)
dummy = pd.get_dummies(train_df["Name"])
train_df = pd.concat([train_df, dummy], axis = 1)
train_df = train_df.drop(["Name", "Undefined"], axis = 1)
train_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(14, 14))
sns.heatmap(train_df.astype(float).corr(), annot=True, cmap="BrBG")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz
import graphviz
x = train_df.drop(["Survived"], axis = 1)
data_train, data_test, target_train, target_test = train_test_split(x, 
                                                                    train_df["Survived"],
                                                                    test_size = 0.1)
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth = 3)
clf.fit(data_train, target_train)
dot_data = export_graphviz(clf, 
                out_file=None, 
                feature_names=x.columns,
                class_names = ["No", "Yes"],
                filled = True,
                special_characters = True)
graphviz.Source(dot_data)
```

```python
pre = clf.predict(data_test)
accuracy_score(pre, target_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=25, max_depth = 6)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
scores = cross_val_score(clf, x, train_df["Survived"], cv=10)
print("åæ¬¡åˆ†æ•¸:", scores)
print("å¹³å‡:", np.average(scores))
```
