```python
from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()
df = pd.DataFrame(boston["data"], columns = boston["feature_names"])
df["target"] = boston["target"]
# df.to_csv("boston.csv")
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(14, 10))
sns.heatmap(df.astype(float).corr(), cmap = "PiYG", annot = True)
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, target_train, target_test = train_test_split(boston["data"], boston["target"], test_size = 0.1)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth = 3)
reg.fit(data_train, target_train)
pre = reg.predict(data_test)
pre
```

```python
from sklearn.tree import export_graphviz
import graphviz
dot = export_graphviz(reg, 
                      out_file = None, 
                      feature_names = boston["feature_names"],
                      filled = True,
                      special_characters = True)
graph = graphviz.Source(dot)
graph.render("boston")
graph
```

```python
import numpy as np
np.subtract(pre, target_test)
```

```python
from sklearn.metrics import r2_score
r2_score(target_test, pre)
```
