```python
import keras
```

```python
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print("訓練資料特徵維度:", x_train.shape)
print("訓練資料標籤維度:", y_train.shape)
print(y_train)
```

```python
import pandas as pd
a = int(input("請輸入第幾張(0 - 59999):"))
pd.DataFrame(x_train[a])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
print("標籤:", y_train[a])
plt.imshow(x_train[a], cmap="gray")
```

```python
x_train_reshape = x_train.reshape(60000, 784).astype(float) / 255
x_test_reshape = x_test.reshape(x_test.shape[0], 784).astype(float) /255
x_train_reshape[0]
```

```python
from keras.utils import np_utils
y_train_cat = np_utils.to_categorical(y_train)
y_test_cat = np_utils.to_categorical(y_test)
print("One-hot後:", y_train_cat[0])
print("One-hot前:", y_train[0])
```

```python
from keras.models import Sequential
from keras.layers import Dense
# Keras三大步驟: 1. model 2. compile 3. evaluate
model = Sequential()
# input_dim只有第一層才要加, 因為keras不知道你的input長啥樣
h1 = Dense(units = 256, input_dim = 784, activation = "relu", kernel_initializer="random_normal")
model.add(h1)
# units > 1 -> softmax, units = 1 -> sigmoid
o = Dense(units = 10, activation = 'softmax', kernel_initializer = 'random_normal')
model.add(o)
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.summary()
```

```python
train_history = model.fit(x = x_train_reshape, 
                          y = y_train_cat,
                          epochs = 10,
                          batch_size = 200,
                          validation_split = 0.1,
                          verbose = 2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(train_history.history['loss'])
plt.plot(train_history.history['val_loss'])
plt.legend(["Loss", "Val Loss"], loc="upper right")
```

```python
# predict_classes -> sklearn(predict)
pre = model.predict_classes(x_test_reshape)
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test, pre))
```

```python
# evaluate -> sklearn(pre -> accuracy_score)
model.evaluate(x_test_reshape, y_test_cat)
```
